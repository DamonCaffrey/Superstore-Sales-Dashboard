```python
temp = list(df_raw["sales"])

for i in range(len(temp)):
    val = temp[i]
    
    if len(val) == 5:
        val = val[0] + val[2:]
    elif len(val) == 6:
        val = val[0:2] + val[3:]

    temp[i] = val

df_raw["sales"] = temp

print(df_raw.shape)
df_raw.head(5)
```

```python
# df_raw.drop(columns=["product_id", "product_name", "customer_name"], inplace=True)
```

```python
def func_replace(col):
    
    df_temp = df_raw[col]
    li_val = []
    
    for val in df_temp:
        
        while '"' in val:
            idx = val.index('"')
            val = val[:idx] + val[idx + 1:]
            
        while "'" in val:
            idx = val.index("'")
            val = val[:idx] + val[idx + 1:]
        
        li_val.append(val)
    return li_val
```

```python
df_raw["ship_mode"] = func_replace("ship_mode")

df_raw["segment"] = func_replace("segment")

df_raw["state"] = func_replace("state")

df_raw["country"] = func_replace("country")

df_raw["region"] = func_replace("region")

df_raw["category"] = func_replace("category")

df_raw["sub_category"] = func_replace("sub_category")

df_raw["order_priority"] = func_replace("order_priority")

df_raw["market"] = func_replace("market")

df_raw["product_id"] = func_replace("product_id")

df_raw["product_name"] = func_replace("product_name")

df_raw["customer_name"] = func_replace("customer_name")
```

```python
df_raw["order_date"] = pd.to_datetime(df_raw["order_date"], dayfirst=True).dt.date
df_raw["ship_date"] = pd.to_datetime(df_raw["ship_date"], dayfirst=True).dt.date
df_raw["sales"] = pd.to_numeric(df_raw["sales"])

print(df_raw.shape)
df_raw.head(5)
```

```python
df_raw.describe(include="all")
```

```python
def func_length(col):
    df_temp = df_raw[col]
    
    lmax = 0
    for v in df_temp:
        l = len(v)
        if lmax < l:
            lmax = l
    return lmax
```

```python
li_columns = list(df_raw.columns)
dict_columns = {}

for col in li_columns:
    
    print(col)
    if isinstance(df_raw[col].iloc[0], np.int64):
        dict_columns[col] = "INT"
    elif isinstance(df_raw[col].iloc[0], np.float64):
        dict_columns[col] = "FLOAT"
    elif isinstance(df_raw[col].iloc[0], str):
        dict_columns[col] = "VARCHAR(" + str(func_length(col) * 2) + ")"
    elif isinstance(df_raw[col].iloc[0], dt.date):
        dict_columns[col] = "DATE"
```

```python
dict_columns
```

```python
s = f"CREATE TABLE orders(\n"
l = len(dict_columns)

for i,key in zip(range(l), dict_columns.keys()):
    
    s += f"{key}\t\t{dict_columns[key]}"
    
    if i == l - 1:
        s += ");"
    else:
        s += ",\n"

print(s)
```

```python
s = 'INSERT INTO orders VALUES\n'
l = len(dict_columns)

start = 50000
end = 51290

for idx in range(start, end):
    
    ser = df_raw.iloc[idx]
    s += '('
    
    for i, key in zip(range(l), dict_columns.keys()):
        
        if dict_columns[key] == 'INT' or dict_columns[key] == 'FLOAT':
            s += f'{ser[i]}'
        else :
            s += f"'{ser[i]}'"
        
        if i < l - 1:
            s += ', '
        else:
            s += ')'
    
    if idx < end - 1:
        s += ',\n'
    else:
        s += ';'
        
print(s)
```
